```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11)
y = 2 * x + 3  # persamaan y = 2x + 3

plt.plot(x, y)
plt.title("Grafik Fungsi Linear")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
