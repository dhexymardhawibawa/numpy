```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 11)
y = 2 * x + 3  # persamaan y = 2x + 3

plt.plot(x, y)
plt.title("Grafik Fungsi Linear")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
angka = np.arange(1, 31).reshape(5, 6)

baris_ketiga = angka[2, :]
kolom_kedua = angka[:, 1]

rata_rata = angka.mean()
maksimum = angka.max()
minimum = angka.min()

angka, baris_ketiga, kolom_kedua, rata_rata, maksimum, minimum
```
